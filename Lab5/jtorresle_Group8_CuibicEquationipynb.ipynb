# Explicación
La función solve_cubic resuelve una ecuación cúbica de la forma $x^3 + bx^2 + cx + d = 0$ por medio de la fórmula cúbica general explicada en el wikipedia de arriba, sin embargo, aquí mostraremos cómo pasamos ésta explicación a un algoritmo.
Primero los coeficientes $b,c$ y $d$ provenientes de la cúbica a resolver se pasan como argumentos a la función.

Posteriormente se calculan p y q que en wikipedia son nombrados como Δ1 y Δ2 los cuales se calculan como $Δ1 = b^2 -3*a*c $ y $Δ2 = 2*b^3 - 9*a*b*c + 27*a^2*d$.

Luego calculamos C1 y C2 que se usrán posteriormente, al igual que l.

Luego procedemos a comprobar si p y q son igual a cero pues de ser ese el caso el método en wikipedia nos indica que el resultado de las tres raíces es $\frac{-b}{3}$.

De no darse el caso anterior debemos proceder a usar el resultado de C1 o C2 pues en sí mismas son la misma fórmula solo que con una diferencia de signo entre ellas, el porqué esta elección es algo arbitrario se debe a que como se explica en wikipedia cuando el resultado de una da 0 debemos simplemente usar el otro C que tiene el signo contrario,por ello la elección de comprobar qué C da igual a cero es totalmente arbitrario, y en este caso elegimos verificar si C1(fórmula con signo positivo) es 0.



```python
import numpy as np
import matplotlib.pyplot as plt
def cubic(b,c,d,x):
    """Calulate the value of the  qubic equation, v = x**3 + bx**2 + c*x +d
    Parameters
    ----------
    b, c, d  : float
       Coefficients

    Returns
    -------
    v : float """

    # ... enter your code here ...
    v = x**3+b*x**2+c*x+d
    return v

x = np.arange(-3.5, 2, 0.001)
z = np.zeros(len(x))
y = cubic(2,3,-2,x)
```

```python
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='$Ecuación Cúbica: x^3 + 2x^2 + 3x - 2$')
plt.plot(x, z, 'r--', label='$y = 0$')  # Línea horizontal y=0 en rojo punteada
plt.xlabel('x')
plt.ylabel('y')
plt.title('Gráfica de una Ecuación Cúbica')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.legend()
plt.show()
```

```python
import cmath

def solve_cubic(b, c,d):
    """Solve a cubic equation, x**3 + bx**2 + c*x +d = 0.

    Parameters
    ----------
    b, c,d : float
       Coefficients

    Returns
    -------
    x1, x2,x3 : float or complex
       Roots.  """

    # ... enter your code here ...
    p = b**2-3*c
    q = 2*b**3-9*b*c+27*d
    C1 = ((q + cmath.sqrt(q**2 - 4*p**3))/2)**(1/3)
    C2 = ((q - cmath.sqrt(q**2 - 4*p**3))/2)**(1/3)
    l = (-1 + cmath.sqrt(-3))/2
    if p == 0 and q == 0:
      x1 = -b/3
      x2 = -b/3
      x3 = -b/3
    elif C1 == 0:
      x1 = -(1/3) * (b + l**0*C2 + p/(l**0*C2))
      x2 = -(1/3) * (b + l**1*C2 + p/(l**1*C2))
      x3 = -(1/3) * (b + l**2*C2 + p/(l**2*C2))
    else:
      x1 = -(1/3) * (b + l**0*C1 + p/(l**0*C1))
      x2 = -(1/3) * (b + l**1*C1 + p/(l**1*C1))
      x3 = -(1/3) * (b + l**2*C1 + p/(l**2*C1))

    return x1,x2,x3
```
